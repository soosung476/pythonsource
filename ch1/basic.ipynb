#### 인덱싱(indexing), 슬라이싱(slicing)
- 인덱스는 0번 부터 시작

```python
a = "Life is too short, You need Python"
print(a[0])
print(a[-1])
print(a[33])
```

```python
# 슬라이싱 -> [start:end] -> start <= index < end, end는 포함하지 않는다.
print(a[0:5])
print(a[0:2])
print(a[19:33])
print(a[:])
print(a[19:])
print(a[:17])
print(a[1:17:2]) # 1부터 17까지 2칸씩 건너뛰면서 가져오기
```

```python
a = "20260714Rainy"
# 분리
print(a[:-5])
print(a[-5:])
year = a[:4]
month = a[4:6]
day = a[6:8]
print(year)
print(month)
print(day)
```

#### 문자열 포매팅
- 문자열 안에 어떤 값을 삽입하는 방법

| 코드 | 설명 |
| --- | --- |
| `%s` | 문자열 String |
| `%c` | 문자 Character |
| `%d` | 십진수 Decimal |
| `%i` | 정수 Integer |
| `%f` | 부동 소수 Floating-point |
| `%b`| 2진수, binary |
| `%o` | 8진수 |
| `%x` | 16진수 |
| `%%` | %문자 자체 |

- format()함수
- f 문자열 포매팅 (After Python 3.6 ~)


```python
a = "I eat %i apples" %-1
a
```

```python
# %s = string 문자열, 어떤 값이든 대응 가능, d = decimal, i = integer, f = float, c = character

a = "I eat %s apples" %'green'
a
```

```python
a = "Err is %d%%" %98
a
```

```python
a = "I ate %d apples. so I was sick for %s days." %(10, "three")
a
```

```python
print("%10s" % "hi") # 10칸 확보하고 그 내에서 오른쪽 정렬
print("%-10sjane" % "hi") # 10칸 확보하고 그 내에서 왼쪽 정렬

print("%0.4f" % 3.42134234) # 소수점 4자리까지 표시
print("%10.4f" % 3.42134234) # 10칸 중 4칸은 소수점, 나머지 6칸은 정수 부분과 공백으로 채워짐
```

```python
a = "I eat {0} apples.".format(5)
a
```

```python
a = "I ate {0} apples. so I was sick for {1} days." .format(10, "three")
a
```

```python
name = "홍길동"
age = 30

# 첫 번째 방법
print("나의 이름은 %s 이고, 나이는 %d 입니다." %(name, age))
# 두 번째 방법
print("나의 이름은 {0} 이고, 나이는 {1} 입니다." .format(name, age))
# f-string 포매팅
print(f"나의 이름은 {name} 이고, 나이는 {age} 입니다.")
```

#### 문자열 함수


```python
a = "hobby"

# 문장열 안에 특정 문자가 몇 번 나오는지 확인
print(a.count('b'))

"test".count('t')
```

```python
# 위치 찾기
a = "Python is the best choice"
a.find("b") # b가 나오는 위치(인덱스), 없으면 -1 반환.
```

```python
a.index("b") # b가 나오는 위치(인덱스), 없으면 오류(ValueError) 발생.
```

```python
# join() : 문자열 삽입


"/".join("abcde")
```

```python
# 대소문자 변경
print("hi".upper())
print("HI".lower())
```

```python
# 공백제거
print("  hi  ".strip())

# 왼쪽 공백 제거
print("  hi  ".lstrip())

# 오른쪽 공백 제거
print("  hi  ".rstrip())
print("hi" == "    hi    ".strip())
```

```python
# 문자열 변경 replace("", "") : 첫 번째 문자열을 두 번째 문자열로 변경
a = "Life is too short"
print(a.replace("Life", "Your leg"))
print(a)
```
