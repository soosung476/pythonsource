```python
# abs(x) : 절대값 x 리턴

print(abs(-3))
print(abs(3))
print(abs(3.12))
```

```python
# chr() <=> ord()
# chr(): 유니코드 값을 넣으면 그 코드에 해당하는 문자를 리턴
# ord(): 문자를 넣으면 그 문자에 해당하는 유니코드를 리턴

print(chr(97))
print(chr(44032))

print(ord('a'))
print(ord('가'))
```

```python
# dir() : 객체가 지닌 변수나 함수를 보여주는 함수

list1 = [1,2,3,]
dir(list1)

import math
dir(math)
```

```python
# divmod(a, b): 몫과 나머지 튜플로 리턴
quot, rem = divmod(7,3)
print(quot)
print(rem)
```

```python
# enumerate(): 리스트, 튜플, 문자열을 입력받아 인덱스 값을 포함하는 enumerate 객체 리턴

for i, name in enumerate(['body','foo','bar'],start=1):
    print(i, name)

for i, name in enumerate('bodybar',start=0):
    print(i, name)
```

```python
# max(), min()
print(max([1,2,3]))
print(max('python'))
print(min([1,2,3]))
print(min('python'))
```

```python
# round(숫자, [, ndigits]): 반올림
print(round(4.6))
print(round(4.678,2))
```

```python
# sorted(): 원본을 건드리지 않고, 입력 데이터를 정렬 후 결과를 새로운 리스트로 리턴

list1 = [1,3,2,6,5]
# 원본 자체를 정렬하는 list1.sort()와는 다름
# list1.sort()
print(sorted(list1))
print(list1)

```

```python
# zip(): 동일한 개수로 이루어진 시퀀스들을 인덱스 끼리 튜플로 묶어서 리스트로 리턴

print(list(zip([1,2,3],[4,5,6])))
print(list(zip([1,2,3],[4,5,6],[7,8,9])))
print(list(zip("abc","def")))
```

```python
# 학생 5명의 점수를 입력받아 다음 정보를 출력하는 프로그램을 작성하시오.

try:
    scores = list(map(int, input("5명의 점수를 입력해주세요(띄어쓰기로 구분됩니다.)").split()))
except:
    print("점수가 잘못 입력되었습니다.")
else:

    print(f"최고점수 {max(scores)}\n최저점수 {min(scores)}\n평균: {sum(scores)/len(scores):.2f}\n오름차순 :{sorted(scores)} ")
    for i,v in enumerate(sorted(scores, reverse=True), start=1):
        print(f"{i}등 : {v}")
# 출력내용
# 최고점수(max), 최저점수(min), 평균점수(sum()/len())(.f2), 점수를 오름차순(sorted)으로 출력, 등수(enumerate)와 함께 출력


```

```python
# 문자열을 입력받아 다음 정보를 출력
# 출력내용: 입력한 문자, 각 문자의 유니코드 값, 가장 큰 문자, 가장 작은 문자
words = input("영어 문자열 입력")

print(words)
for i in words:
    print(f"{i} -> {ord(i)}")
# 문자열 입력 : Python
# P -> 80
# y -> 121
print(f"가장 큰 문자 {max(words)}")
print(f"가장 작은 문자 {min(words)}")

# 가장 큰 문자: y
# 가장 작은 문자: P
```

```python
x = [3,7,1,-5,9]
y = [2,-4,8,10,-1]

coo = (list(zip(x,y)))

for i,(x, y) in enumerate(coo,start=1) :
    print(f" {i}번 좌표: ({x},{y}) -> {abs(x)+abs(y)}")
# zip, abs, enumerate()
# 출력
# (3,2) -> 5
```

```python
# 사용자로부터 숫자 10개 입력받기

class MyError(Exception):
    def __str__(self):
        return "숫자를 10개만 입력해주세요"
try:
    number = list(map(int, input("숫자를 10개 입력해 주세요 (띄어쓰기로 구분됩니다.)").split()))
    if(len(number) !=10):
        raise MyError
except:
    print("조건에 맞게 입력해주세요")
else:
    print(f"원본 리스트 {number}")
    print(f"중복 제거 {set(number)}")
    print(f"오름차순 정렬{sorted(set(number))}")
    print(f"리스트 -> 튜플{tuple(sorted(set(number)))}")
    print(f"중복 제거 후 집합 크기{len(set(number))}")
# 출력
# 전체 입력 리스트, 중복 제거, 오름차순, 튜플변환, 집합 크기

# 3,5,2,5,7
# 원본 [3,5,2,5,7]
# 중복제거 {1,2,4,7}
# 정렬
# 튜플
# 집합 크기: len()
```
