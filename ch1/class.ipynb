###

| 구분 | 첫 인자 | 접근 가능| 주 용도 |
|----|----|----|----|
| `인스턴스 메서드` | `self` | 인스턴스 변수, 클래스 변수 모두 | 객체별 동작 |
| `@classmethod` | `cls` | 클래스 변수만 | 클래스 단위 동작 |
| `@staticmethod` | 없음 | 없음(매개변수로 받은 것만) | 클래스와 관련은 있지만 독립적인 계산 |


```python
class Student():
    # 클래스 변수
    school_name = "Soldesk"
    student_count = 0

    def __init__(self,name):
        self.name = name
        Student.student_count += 1

    # 인스턴스 메서드    
    def show_info(self):
        print("=== 인스턴스 메서드 ===")
        print(f"학생이름 : {self.name}")    # 인스턴스 변수 접근
        print(f"학교명 : {Student.school_name}")    #클래스 변수 접근
        print(f"학생수 : {Student.student_count}")


    @classmethod
    def show_school_info(cls):
        print(f"학교명 : {cls.school_name}")
        print(f"학생수 : {cls.student_count}")


    @staticmethod
    def add_score(score1, score2):
        print(f"총점 : {score1 + score2}")



stu1 = Student("홍길동")
stu2 = Student("성춘향")

# 인스턴스 매서드
stu1.show_info()
stu2.show_info()

# 클래스 메서드
print()
Student.show_school_info()


# 정적 메서드
print()
Student.add_score(80,95)
```

```python
# Account 클래스 작성
# 멤버변수 : 계좌주(owner),  잔액(balance)
# 멤버메소드 : deposit(money) = 잔액 +입금액 , 
#             withdraw() 잔액 -출금액
#             showBalance() : 현재 잔액 출력

class Account:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance
    def deposit(self, money):
        self.balance += money
    def withdraw(self, money):
        if self.balance < money:
            print("잔액 부족")
        else:
            self.balance -= money
    def showBalance(self):
        print(f"현재 잔액 : {self.balance}")


a1 = Account("kim", 250000)
a2 = Account("park", 7300000)

a1.showBalance()
a1.deposit(30000)
a1.showBalance()
a1.withdraw(20000)
a1.showBalance()

    
```

```python
# Book 클래스 작성
# 클래스 변수 : book_count
# 멤버변수 : 제목(title), 저자(author)
#   메소드: showInfo() => 제목 : ## 
#                        저자 : ##
# 현재 책 권수 출력 


class Book:
    book_count = 0

    def __init__(self, title, author):
        self.title = title
        self.author = author
        Book.book_count += 1
        
    def showInfo(self):
        print(f"제목: {self.title}")
        print(f"저자: {self.author}")

    @classmethod
    def count(cls):
        print(f"현재 책 권수는 {cls.book_count} 입니다")

b1 = Book("Python programming", "kim")
b1.showInfo()
Book.count()
```

```python
# Order class 작성
# 클래스 변수: total_order
# 멤버 변수 : 메뉴(menu), 가격(price), 수량(count)
# 메소드 : getTotalPrice() => 가격*수량 : 총 주문 금액 반환
#       : showOrder() : 출력문 작성
#         메뉴 : 아메리카노
#         가격 : 4000원
#         수량 : 2개
#         총 주문금액 : 8000원

class Order:
    total_order = 0
    def __init__(self, menu, price, count):
        self.menu = menu
        self.price = price
        self.count = count
        Order.total_order += 1

    def getTotalPrice(self):
        return int(self.price)*int(self.count)

    def showOrder(self):
        print(f"메뉴 : {self.menu}")
        print(f"가격 : {self.price}")
        print(f"수량 : {self.count}")
        print(f"총 주문금액 : {self.getTotalPrice()}원")


o1 = Order("아메리카노", 2000, 3)
o2 = Order("국밥", 12000, 2)

o1.showOrder()
print()
o2.showOrder()

print(f"전체 주문 수 : {Order.total_order}")
```

#### 상속
- 기존 클래스(부모/상위)의 멤버변수와 메서드를 새로운 클래스(자식/하위)가 그대로 물려받는 것
- 코드 재사용성, 확장

```python
class Parent:
    def __init__(self):
        self.value = "test"
        print("Parent 클래스의 __init__() 호출")

    def test(self):
        print("parent 클래스의 test() 호출")

class Child(Parent):
    pass

c1 = Child()
c1.test()
print(c1.value)
```

```python
# super() = superior class

class Child(Parent):
    def __init__(self):
        super().__init__()
        print("Child 클래스의 __init__() 호출")

c1 = Child()
c1.test()
print(c1.value)
           
```

```python
# Audio : 멤버변수 : power, volume / 메소드 : turn_on() turn_off(), adjust_volume() + tune()
# TV : 멤버변수: power, volume, size / 메소드 : turn_on() turn_off(), adjust_volume() + watch()


class AudioVisual:
    def __init__(self,power,volume):
        self.power = power
        self.volume = volume

    def switch(self, on_off):
        self.power = on_off

    def setVolume(self, volume):
        self.volume = volume


class Audio(AudioVisual):
    def __init__(self, power, volume):
        super().__init__(power, volume)
    def tune(self):
        print("La la la..." if self.power else "turn it on")

class TV(AudioVisual):
    def __init__(self, power, volume,size):
        super().__init__(power, volume)
        self.size = size

    def watch(self):
        print("have fun!" if self.power else "switch on")



tv = TV(False,12,50)
tv.switch(True)
tv.watch()


audio = Audio(True, 15)
audio.setVolume(10)
audio.tune()
audio.switch(False)
```

#### 오버라이딩
- 부모 클래스에 있는 메서드를 자식 클래스에서 같은 이름으로 다시 정의 해서 동작 내용 변경

```python
class Animal:
    def __init__(self, name):
        self.name = name
    def sound(self):
        print(f"{self.name}가 소리를 냅니다")


class Dog(Animal):
    def __init__(self, name):
        super().__init__(name)

    def sound(self):
        print(f"{self.name}: ruff! ruff!")

class Cat(Animal):
    def __init__(self, name):
        super().__init__(name)

    def sound(self):
        # 오버라이딩 시 동일한 부모의 메소드를 호출하고 싶다면
        super().sound()
        
        print(f"{self.name}: meow meow!")


dog = Dog("Tipsy")
cat = Cat("Moly")

dog.sound()
cat.sound()
```

```python
# Employee 부모 클래스 :
# 멤버 변수 : 이름(name), 급여(pay)
#     메소드: showInfo() => 이름과 급여 출력

# Manager 자식 클래스
# 멤버 변수 : 보너스(bonus)
# 메소드 : getTotalPay() :급여 + 보너스 출력


class Employee:
    def __init__(self, name, pay):
        self.name = name
        self.pay = pay

    def showInfo(self):
        print(f"name: {self.name}, pay: {self.pay}")

class Manager(Employee):
    def __init__(self, name, pay, bonus):
        super().__init__(name, pay)
        self.bonus = bonus

    def getTotalPay(self):
        print(f"total payment : {self.pay + self.bonus}")
        

emp = Employee("홍길동", 200)
emp.showInfo()

manager1 = Manager("sarah",350,50)
manager1.showInfo()
manager1.getTotalPay()

```

```python
# Shape 부모 클래스
# 메서드 : area() => "도형의 넓이" 출력

# Rectangle 자식 클래스
# 멤버 변수 : width, height
# 메서드 : area() => 직사각형의 넓이 출력

# Circle 자식 클래스
# 멤버 변수 : 반지름(radius)
#    메서드 : area() => 원의 넓이 출력

class Shape:
    def area(self):
        print("도형의 넓이")


class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        print(f"직사각형의 넓이: {self.width * self.height}")

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        print(f"원의 넓이: {self.radius**2*3.141592 :.2f}")

rectagle1 = Rectangle(12,12)
rectagle1.area()

circle1 = Circle(2)
circle1.area()
```
