#### 클래스
- 설계도
- 클래스명은 대문자로 시작
- 메소드(== 클래스 내부의 함수) 


```python
class Calculator:
    def __init__(self):
        self.result = 0

    def add(self, num):
        self.result += num
        return self.result

    def sub(self, num):
        self.result -= num
        return self.result
```

```python
# 객체 생성
# 변수 = 클래스명()

calc1 = Calculator()
print(calc1.add(3))
print(calc1.add(4))


calc2 = Calculator()
print(calc2.add(13))
print(calc2.add(14))
```

```python
class Student:
    def __init__(self,name,number,grade,detail):
        self.name = name
        self.number = number
        self.grade = grade
        self.detail = detail
        pass

# 인스턴스를 출력할 때 출력시 원하는 값들을 반환
    def __str__(self):
        return f"이름: {self.name}, 반: {self.number}, 학번: {self.grade}, 상세정보: {self.detail}"
    

```

```python
# 객체 생성 => __init__() 자동으로 호출됨 , 생성자.
s1 = Student("kim",1,2,{"gender":"male","score1":95,"score2":88})
s2 = Student("park",2,2,{"gender":"female","score1":75,"score2":68})
s3 = Student("hong",2,1,{"gender":"male","score1":99,"score2":86})

# s1은 Student 클래스의 인스턴스임
print(s1) #<__main__.Student object at 0x000001396C2C2D70>
print(s2)
print(s3)
```
