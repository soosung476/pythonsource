`숫자가 짝수면 짝수라고 출력, 아니면 홀수라고 출력`

```python
num = 5


if num%2 == 0:
    print("짝수")
else:
    print("홀수")
```

```python
# >, <, >=, <=, ==, !=, in, not in, is : 비교연산자 => return값은 True, False
a = 200
# a가 100보다 작다면 a는 100보다 작다 출력

if a < 100:
    print("a는 100보다 작다")
elif 100 < a <=200:
    print("a는 100과 200사이의 수")
else:
    print("a는 200보다 크다")
```

```python
# 점수가 90 이상이면 A, 80이상이면 B, 70 이상이면 C, 나머지는 F

score = 99

if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
else:
    print("F")

```

```python
a,b,c = 12,6,19

# 가장 큰 숫자 출력
max = a
if max <b:
    max = b

if max <c:
    max = c

print(max)
```

```python
# 가장 작은 수

min = a
if min > b:
    min = b;
if min > c:
    min = c;

print(min)
```

```python
# print sep, end 사용법
print(a,b,c)
print(a,b,c, sep=", ", end=" ")
print(b)

print("T","E","S","T", sep="")
```

```python
fruits =['딸기','사과','바나나','배','수박']

# fruits 리스트에 딸기라는 과일이 포함되어 있다면, 딸기가 포함된 출력, 아니면 딸기가 없음 출력

if '딸기' in fruits:
    print("딸기가 포함됨")
else :
    print("딸기가 없음")
```

```python
# score가 90 이상이고 grade가 A인 경우 합격 나머지는 불합격


score, grade = 85,'A'

if(score >=90 and grade =='A'):
    print("합격")
else:
    print("불합격")
```

```python
# score가 90 이상이거나 grade가 A인 경우 합격 나머지는 불합격

score, grade = 85,'A'
if score >= 90 or grade =='A':
    print("합격")
else:
    print("불합격")
```
