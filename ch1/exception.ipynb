```python
# 예외 클래스 생성 가능
class MyError(Exception):
    def __str__(self):
        return "허용되지 않는 별명입니다."

def say_nick(nick):
    if nick == "바보":
        raise MyError()
    print(nick)



try:
    say_nick("천사")
    say_nick("바보")
except MyError as e:
    print(e)
    
```

```python
import random

class MyError(Exception):
    def __str__(self):
        return "비밀번호의 자릿수를 입력해주세요(4 이상)."

chars = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789!@#$%^&*()"

try:
    number_length = int(input("비밀번호의 길이를 알려주세요 (4자리 이상)"))
    if (number_length) <4:
        raise MyError
except MyError as e:
    print(e)
else:
    pwd = random.choices(chars, k = number_length)
#리스트 => 문자열로 변경
    print(f"비밀번호는: {"".join(pwd)}")
    
# 문제 1. 랜덤 비밀번호 생성기
# 사용자로부터 생성할 비밀번호의 길이를 입력받아 랜덤 비밀번호를 생성하는 프로그램 작성
# choice()

# 비밀번호 길이는 4 이상만 허용
# 숫자가 아닌 값을 입력하면 예외 처리


```

```python
# 나이 계산 프로그램
# 생년월일을 입력받아 현재 나이를 계산하는 프로그램 작성
# 입력형식 2001-08-25
# input() => 2001-08-25 => 날짜형식
# datetime.datetime.striptime("2001-08-25,","%Y-%m-%d")
import datetime
dob = input("생년월일을 입력해 주세요 ex) 2002-08-25")

try:
    dob = datetime.datetime.strptime(dob,"%Y-%m-%d")
    
except ValueError as e:
    print(e)
else:
    today = datetime.datetime.now()
    age = today.year - dob.year

    if(today.month, today.day ) < (dob.month, dob.day):
        age -= 1
    
    print(age)
```

```python
# 학생 정보를 json으로 저장하기
# 학생정보를 입력받아 json 형식으로 파일 저장
# 이름, 나이, 전공
# 읽어서 출력
import json

class MyError(Exception):
    def __str__(self):
        return "모든 정보를 전부 입력해주세요"


student_info1 = list(input("학생 정보를 입력해주세요 ex) 홍길동, 25, 물리학").split(","))
keys = ["이름","나이","전공"]
student_info = []
try :
    for j in student_info1:
        student_info.append(j.strip())
    if len(student_info) <3:
        raise MyError
except MyError as e:
    print(e)
else:

    dict1 = {keys[0]:student_info[0], 
        keys[1]:student_info[1], 
        keys[2]:student_info[2]}

with open("./data/studentinfo1.json","w",encoding="utf-8") as f:
    json.dump(dict1, f, ensure_ascii=False)



with open("./data/studentinfo1.json","r",encoding="utf-8") as f:
    data = json.load(f)
    print(data)

```

```python
# 다음과 같은 csv 파일이 있다.
# 이름, 국어, 영어, 수학
# 홍길동, 90, 88, 70
# 김철수, 75, 80, 92
# 이영희, 77, 69, 86
# 성춘향, 88, 75, 95
import csv

with open("./data/sample5.csv","w",encoding="utf-8-sig",newline="") as f:
    headers = ["이름", "국어", "영어", "수학"]
    writer = csv.DictWriter(f, fieldnames=headers)
    writer.writeheader()
    writer.writerow({"이름": "홍길동","국어":90,"영어":88,"수학":70})
    writer.writerow({"이름": "김철수","국어":75,"영어":80,"수학":92})
    writer.writerow({"이름": "이영희","국어":77,"영어":69,"수학":86})
    writer.writerow({"이름": "성춘향","국어":88,"영어":75,"수학":95})


with open("./data/sample5.csv","r",encoding="utf-8-sig") as f:
    reader = csv.DictReader(f)
    try:
        for student in reader:
            average = (
                int(student["국어"])
                + int(student["영어"])
                + int(student["수학"])
            ) / 3

            print(f'{student["이름"]} : {average:.2f}')
    except KeyError as e:
        print("KeyError has occured")
    
# csv 파일을 읽어 학생별 평균 점수 출력
# ex) 홍길동 : 80.0

#csv.DictReader() 사용, 평균은 둘째자리까지
```
