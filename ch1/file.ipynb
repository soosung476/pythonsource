#### file 읽고 쓰기

| 파일 열기 모드 | 설명 |
|--------------|------|
|`r`| **읽기(Read)모드** : 파일의 내용을 읽기만 할 때 사용. 파일이 존재하지 않는 경우 오류 발생.|
|`w`| **쓰기(Write)모드** : 파일의 내용을 새로 쓸때 사용. 기존 파일이 존재하는 경우 내용 삭제 후 새로 작성. 없는 경우 새로 생성됨.|
|`a`| **추가(Append)모드** : 기존 파일의 마지막에 내용을 추가할 때 사용. 파일이 없는 경우 새로 생성.|

```python
# sample.txt 파일을 읽어서 총합과 평균을 구한 후 총합과 평균을 result.txt 파일에 작성

samples = []
with open('./data/sample.txt','r',encoding='utf-8') as f:
    for num in f:
        samples.append(int(num))


with open('./data/result.txt','w',encoding='utf-8') as f:
    f.write(f"총합 : {sum(samples)}\n평균 : {sum(samples)/len(samples):.2f}")
```

```python
# fruits.txt 읽어서 단어의 개수 출력
# apple 3
# banana 2

with open('./data/fruits.txt', 'r', encoding='utf-8') as f:
    contents = f.read().split()

dict1 = {}
for k in contents:
    if k in dict1:
        dict1[k] += 1
    else :
        dict1[k] = 1

for fruit, count in dict1.items():
    print(f"{fruit} : {count}")
```

```python
# 1000명의 이름, 키, 몸무게 test1.txt 읽은 후
# BMI 지수 계산
# bmi = 체중 / 키**2
# bmi가 18.5 미만일 때 저체중
# bmi가 18.5 ~ 22.9 정상
# bmi가 23 ~ 24.9 과체중
# bmi가 25 비만


with open('./data/test1.txt', 'r', encoding='utf-8') as f:
    for line in f:
        name, weight, height = (f.readline().rstrip()).split(", ")
        bmi = int(weight)/ ((int(height)/100)**2)
        result = ""
        if bmi < 18.5:
            result = "저체중"
        elif 18.5 < bmi < 22.9:
            result = "정상"
        elif 22.9 < bmi < 24.9:
            result = "과체중"
        else : 
            result = "비만"
        print("\n".join([f"이름 : {name}\n몸무게 : {weight}\n키 : {height}\nBMI : {bmi:.1f}\n결과 : {result}"]))
        print()
```

```python
# origin.txt 읽어서 암호화 시킨 후 encr.txt 저장
# chr(ord("str"+100))
# 1. 암호화 선택 : origin.txt 읽어서 변수에 담기 그걸 encr.txt에 쓰기
# 2. 복호화 선택 : 암호화된 파일을 읽은 후 원래대로 출력

while(1):
    no = int(input("1.암호화 | 2. 복호화 | 3. 종료 | 번호를 입력하세요 "))
    if no == 1:
        with open('./data/origin.txt','r',encoding='utf-8') as reader:
            with open('./data/encr.txt','w',encoding='utf-8') as writer:
                for line in reader:
                    for i in range(len(line)):
                        writer.write(chr(ord(line[i])+100))
                
    elif no == 2:
        with open('./data/encr.txt','r',encoding='utf-8') as f:
            for line in f:
                for i in range(len(line)):
                    print(chr(ord(line[i])-100), end="")
    else :
        break

    break
    
```
