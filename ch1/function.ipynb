````python
# add
# add(2,3)
# add(2,3,4,5)
# add(2,3,4,5,6,7)

# 가변 파라미터
# 입력값이 몇 개가 될 지 모르는 경우
# *

# def 함수명(*변수명):
#     해야할 일

def add(*args):
    result = 0
    for i in args:
        result += i
    print(result)
    return result
````

````python
add(2,3)
add(2,3,4,5)
add(2,3,4,5,6,7)
````

````python
# def add_mul(choice, *args, a):
#     pass


# 일반 매개변수는 앞 쪽으로 선언
# +, * 실행
def add_mul(choice, *args):
    result = 0;
    if choice =="add":
        for i in args:
            result += i

    elif choice == "mul":
        result = 1;
        for i in args:
            result *= i
    else:
        print("choose add or mul")
    return result
````

````python

# add_mul("add",10,11,12,13, a= 3)
print(add_mul("add",10,11,12,13))
print(add_mul("mul",10,11,12,13))
````

````python
# 키워드 매개변수 (key =value)
# **변수명

# 딕셔너리 , (key=value)
def print_kwargs(**kwargs):
    print(kwargs)

    # key값만 출력
    for k in kwargs.keys():
        print(k)

    # value만 출력
    for v in kwargs.values():
        print(v)
    # key, value 둘 다 출력
    for i in kwargs.items():
        print(i)
    
````

````python
print_kwargs(name ="kim", age =15)
print_kwargs(name ="park", age =25, addr = "서울")
print_kwargs(name ="hong", age =30, addr = "경기", tel = "010-1234-5678")
````

````python
# 변수의 효력 범위

# 함수 밖의 변수 : 전역변수 global variable
a = 1
def vartest(a):
    # parameter a : local variable
    a = a+1


vartest(a)
print(a)
````

````python
# how to change global variable?
# 1) use return (recommended)

a = 1
def vartest(a):
    # parameter a : local variable
    a = a+1
    return a

a = vartest(a)
print(a)
````

````python
# 2) use global keyword
# can't access global variable in function generally.
# this way has critical risk, so use method 1)

a = 1
def vartest():
    global a
    a = a+1


vartest()
print(a)
````

````python
# 입력으로 숫자(자연수)를 받아서 자연수가 짝수라면 True를 리턴, 아니면 False를 리턴하는 함수
# is_even() 

def is_even(args):
    if (args%2 == 0):
        return True
    else: 
        return False

print(is_even(125))
print(is_even(124))

# True가 리턴되면 짝수/ False 홀수
# if (is_even(3)):
#     print("짝수")
# else:
#     print("홀수")


print("짝수") if is_even(5) else print("홀수")
````

````python
# 리스트의 평균 구하는 함수
# get_avarage()


def get_avarage(*args):
    result = 0
    for i in args:
        result += i
    print(result/len(args))


get_avarage(10,20)
````

````python
# 사칙연산 함수
# 결과 리턴

def calculator(arg1, arg2, sign):
    if sign == "+":
        return arg1+arg2
    elif sign == "-":
        return arg1-arg2
    elif sign =="*":
        return arg1*arg2
    elif sign == "/":
        return arg1/arg2
    elif sign == "//":
        return arg1//arg2


print(calculator(10,20,"+"))
print(calculator(10,20,"-"))
print(calculator(10,20,"*"))
print(calculator(10,20,"/"))
print(calculator(10,20,"//"))

````

````python
# filter()
list1 = list(range(1,21))

# 짝수만 모아서 새로운 리스트 생성

# 방법 1
even_list = []
for i in list1:
    if i%2 == 0:
        even_list.append(i)


# 방법 2
even_list = [i for i in list1 if i%2 == 0]

# 방법 3
list2 = list(filter(lambda i: i%2==0, list1))
print(list2)
````

````python
strings = ["Bob","charles","Alexander3","teddy","apple","Banana"]

# 각 단어의 시작이 uppercase인 단어들만 모아서 새로운 리스트로 생성

strings2 = []

# using for
for i in strings:
    if i[0].isupper():
        strings2.append(i)
print(strings2)


# using list comprehension
strings2 = [i for i in strings if i[0].isupper()]
print(strings2)


# using filter() with lambda
strings2 = list(filter(lambda x: x[0].isupper(), strings))
print(strings2)
````
