```python
num = (input("숫자 입력"))

# 사용자에게 숫자를 입력받은 후 +10 한 결과 출력
if num.isdigit():
    num =int(num)+10
    print(num)
else:
    print("숫자를 입력하세요")
```

```python
# 사용자 입출력
# input()
# print()

def calculator(arg1, arg2, sign):
    if sign == "+":
        return arg1+arg2
    elif sign == "-":
        return arg1-arg2
    elif sign =="*":
        return arg1*arg2
    elif sign == "/":
        return arg1/arg2
    elif sign == "//":
        return arg1//arg2
    
# 숫자 2개 입력받기, 연산자(+,*,....) 기호 입력받기
num1 = int(input("첫번째 숫자 입력"))
num2 = int(input("두번째 숫자 입력"))
op = input("연산자 입력 (+,-,*,/,//)")


print(calculator(num1,num2,op))
```

```python
values = input("두 숫자와 연산자 기호 입력 ex) 15 10 +")
# result = values.split()
# print(result)
num1,num2,op = values.split()

print(calculator(int(num1),int(num2),op))
```

```python
# 15,10,+
values = input("두 숫자와 연산자 기호 입력 ex) 15,10,+")
num1,num2,op = values.split(",")
print(calculator(int(num1),int(num2),op))
```
