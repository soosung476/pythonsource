```python
import math
# 반올림, 올림, 버림, sin, cos
print(math.cos(math.pi*2))

print(math.sin(0))
print(math.sin(math.pi/2))
print(math.sin(math.pi*2))


print(math.ceil(3.2))
math.tan(0)
```

```python
# import 모듈명
# from 모듈명 import 사용하고 싶은 함수, 클래스, 변수 나열... => 모듈명 제거 가능

# as
# import 모듈명 as 별칭
# from 모듈명 import 사용하고싶은 함수 as 별칭

from math import cos, pi, floor, ceil, sin

print(cos(pi))
print(ceil(3.2224))
print(floor(1.3))
print(pi)
```

```python
import math as m

print(m.cos(pi))
print(m.pi)
print(m.floor(2.76))
print(m.ceil(2.3))
```

```python
from math import sin as s
s(0)
```
