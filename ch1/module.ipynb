```python
import random

print(random.random())
# random() : 0.0 ~1.0 사이의 실수 중에서 난수
print(random.randint(1,10))

print(random.uniform(10,20))
# uniform(최소,최대) : 지정한 두 숫자 실수 사이의 난수

print(random.randrange(10))
# randrange(start, stop, step)

print(random.choice([1,2,3,4,5,6,7]))

list1 = list(range(1,8))
random.shuffle(list1)
# shuffle(list)

print(list1)

# sample(시퀀스, k= 개수, counts =[가중치])
print(random.sample("Python", k=2))
```

```python
from random import random, randint, uniform, randrange, choice, shuffle, sample


print(random())
# random() : 0.0 ~1.0 사이의 실수 중에서 난수
print(randint(1,10))

print(uniform(10,20))
# uniform(최소,최대) : 지정한 두 숫자 실수 사이의 난수

print(randrange(10))
# randrange(start, stop, step)

print(choice([1,2,3,4,5,6,7]))

list1 = list(range(1,8))
shuffle(list1)
# shuffle(list)

print(list1)

# sample(시퀀스, k= 개수, counts =[가중치])
print(sample("Python", k=2))

```

```python
# 쉼표로 구분된 파일
# 혹은 다양한 기호로 분리
import csv

# csv.reader
with open("./data/sample1.csv", "r") as f:
    reader = csv.reader(f)
    # print(reader)
    next(reader)
    for c in reader:
        print(c)
```

```python
with open("./data/sample1.csv", "r") as f:

    # csv.DictReader
    reader = csv.DictReader(f)
    # print(reader)
    # next(reader)
    for c in reader:
        for k,v in c.items():
            print(k,v)
        print()
```

```python

with open("./data/sample4.csv", "w", encoding="utf-8-sig",newline="") as f:
    headers = ['이름', '국어', '영어']
    writer = csv.DictWriter(f, fieldnames=headers)
    # 헤더작성
    writer.writeheader()
    writer.writerow({'이름':'홍길동', '국어':88, '영어':92})
    writer.writerow({'이름':'성춘향', '국어':87, '영어':90})
```

```python
import json

with open("./data/myinfo.json","r",encoding='utf-8') as f:
    data = json.load(f)
    print(data)
    print(type(data))
```

```python
dict1 = {'name': '성춘향', 'birth': '0927', 'age': 24}

with open("./data/myinfo1.json","w",encoding='utf-8') as f:
    json.dump(dict1, f, ensure_ascii=False)
```

```python
# loads(): json 문자열 => 파이썬 자료형(dict)으로 변환


str1 = """
{
    "name": "성춘향", 
    "birth": "0927", 
    "age": 24
}
"""
data = json.loads(str1)
print(data)
print(type(data))

print(data['name'])
print(data.get('name'))
```

```python
# dumps() : 파이썬 자료형(dict) => json 문자열로 변환
dict1 = {
    'name': '홍길동', 
    'birth': '0727', 
    'age': 24
    }


data = json.dumps(dict1, ensure_ascii=False)
print(data)
print(type(data))
```
